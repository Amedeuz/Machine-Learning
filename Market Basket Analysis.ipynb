```python
# Intializing the list
transactions = []

# populating a list of transactions
for i in range(0, 7501): 
    transactions.append([str(df.values[i,j]) for j in range(0, 20)])
```

```python
transactions
```

```python
#use a for loop to remove all zeroes in all the lists
for sub in transactions:
     sub[:] = [ele for ele in sub if ele != '0']
```

```python
#show the second and third transaction without zeroes
transactions[1:3]
```

```python
#import TransactionEncoder
from mlxtend.preprocessing import TransactionEncoder

# Instantiate transaction encoder and identify unique items in transactions
encoder = TransactionEncoder().fit(transactions)

# One-hot encode transactions
onehot = encoder.transform(transactions)

# Convert one-hot encoded data to DataFrame
onehot = pd.DataFrame(onehot, columns = encoder.columns_)

# Print the one-hot encoded transaction dataset
print(onehot)
```

```python
# Compute the support
support = onehot.mean()

# Print the support
print(support)
```

```python
onehot.shape
```

```python
support.shape
```

```python
#load apriori and association_rules from mlxtend
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
#apply the apriori algorithm
frequent_items = apriori(onehot, min_support = 0.01, max_len = 3, use_colnames = True)

#print number of frequent items or rules
print(len(frequent_items))
```

```python
print(frequent_items.head(10))
```

```python
frequent_items[0:5]
```

```python
#print the top 5 support
frequent_items.sort_values(by='support', ascending=False)[0:5]
```

```python
#compute association rules
rules = association_rules(frequent_items, metric='support', min_threshold = 0.03)
```

```python
#print number of rules
print(len(rules))
```
